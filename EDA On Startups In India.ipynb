```python
plt.figure(figsize=(15,5))
df.groupby('Starting Year')['Company'].count().plot.bar()
plt.ylabel("No of companies")
```

```python
df[df['Starting Year']==2015].sort_values(by=['Funding Amount in $'], ascending=False)
```

```python
years=df['Starting Year'].unique()
years=np.sort(years)
```

```python
years
```

```python
df['Starting Year'].nunique()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='Funding Round',y='Company',data=df.sort_values('Funding Round', ascending=False).iloc[:10],alpha=0.8)  
plt.title('Companies with highest funding rounds',fontsize=20)    #Note labelling the data
plt.ylabel('Company',fontsize=20)               #Note labelling the y-label
plt.xlabel('Count',fontsize=20)               #Note labelling the x-label
plt.tick_params(labelsize=8)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='Funding Amount in $',y='Company',data=df.sort_values('Funding Amount in $', ascending=False).iloc[:10],alpha=0.8)  
plt.title('Companies with highest funding ',fontsize=20)    #Note labelling the data
plt.ylabel('Company',fontsize=20)               #Note labelling the y-label
plt.xlabel('Funding Amount in $',fontsize=20)               #Note labelling the x-label
plt.tick_params(labelsize=8)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='No. of Investors',y='Company',data=df.sort_values('No. of Investors', ascending=False).iloc[:10],alpha=0.8)  
plt.title('Companies with highest No. of Investors',fontsize=20)    #Note labelling the data
plt.ylabel('Company',fontsize=20)               #Note labelling the y-label
plt.xlabel('No. of Investors',fontsize=20)               #Note labelling the x-label
plt.tick_params(labelsize=8)
plt.show()
```

```python
fig, axes = plt.subplots(13, 2, sharex=True, figsize=(16,50))
fig.tight_layout(pad=10)
for n in years:
    i=1+ int(np.where(years==n)[0])
    plt.subplot(13,2,i, title=n)
    sns.barplot(y='Company', x='Funding Amount in $', data=df[df['Starting Year']==n].sort_values('Funding Amount in $', ascending=False).iloc[:10])
    plt.xticks(rotation=0)
```
