# Import Libraries

```python
import numpy 
import pandas 
import cv2
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Load image and display it in notebook.

# Load Image

```python
img = cv2.imread('IMG_2160-min.jpg')
```

### Visualization

```python
plt.imshow(img)
```

```python
img_rbg = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rbg)
```

### Flip the image upside down and display in notebook.

```python
new_img = img_rbg.copy()
new_img = cv2.flip(new_img,0)
plt.imshow(new_img)
```

### Draw a empty RED rectangle around the dog face display the image in the notebook. 

```python
pt1 = (180,260)
pt2 = (250,350)
```

```python
cv2.rectangle(img_rbg,pt1=pt1,pt2=pt2,color=(255,0,0),thickness=10)
plt.imshow(img_rbg)
```

### Draw BLUE TRIANGLE in the middle of image and display in the notebook.

```python
vertices = np.array([ [250,300], [200,350], [230,250]],np.int32)
```

```python
vertices.shape
```

```python
pts = vertices.reshape(-1,1,2)
```

```python
pts
```

```python
img_rbg = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.polylines(img_rbg,[pts], color =(0,0,255),isClosed= True,thickness =20)
plt.imshow(img_rbg)
```

### Just filled out TRIANGLE,and show it in notebook.

```python
img_rbg = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.fillPoly(img_rbg,[pts], color =(0,0,255))
plt.imshow(img_rbg)
```

### YOU HAVE TO RUN AS A SCRIPT. Create a script that open the picture and allows to draw an empty circle whenever you click RIGHT MOUSE BUTTON DOWN

```python
import cv2
import numpy as np 

#########################
#### CREATE FUNCTION ###
########################

def create_circle(event,x,y,flags,param):
    if event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img,(x,y),100,(0,0,255),thickness=10)
        

img = cv2.imread('IMG_2160-min.jpg')

cv2.namedWindow(winname='Dog')

cv2.setMouseCallback('Dog',create_circle)

while True:
    
    
    cv2.imshow('Dog',img)
    
    if cv2.waitKey(20) & 0xFF == 27:
        break
        
cv2.destroyAllWindows()        
```
