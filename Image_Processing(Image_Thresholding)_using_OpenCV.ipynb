## Creating Function to show above pic

```python
def show_pic(img2):
    fig = plt.figure(figsize=(12,12))
    ax = fig.add_subplot(111)
    ax.imshow(img2,cmap='gray')
```

```python
show_pic(img2)
```

```python
ret , thre1 = cv2.threshold(img2,127,255,cv2.THRESH_BINARY)
show_pic(thre1)
```

```python
ret , thre1 = cv2.threshold(img2,180,255,cv2.THRESH_BINARY)
show_pic(thre1)
```

## Adopative Thresold

```python
the2 = cv2.adaptiveThreshold(img2,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8)
show_pic(the2)
```

```python
blended = cv2.addWeighted(src1=thre1, alpha=0.6,src2=the2, gamma=0,beta=0.4)
show_pic(blended)
```

```python
blended = cv2.addWeighted(src1=thre1, alpha=0.5,src2=the2, gamma=0,beta=0.4)
show_pic(blended)
```
