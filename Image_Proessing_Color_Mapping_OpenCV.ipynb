# IMPORT LIBRARIES

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# LOAD IMAGE

```python
img = cv2.imread('808683.jpg')
```

## DISPLAY IMAGE

```python
plt.imshow(img) # DUE TO CV2 COLOR CHANNEL IT IS BGR COLOR CHANNEL
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
## HSV - HUE SATURATION VALUE AND HLS - HUE LIGHTNESS SATURATION
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
plt.imshow(img)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2HLS)
plt.imshow(img)
```
